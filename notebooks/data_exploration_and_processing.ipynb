```python
from pathlib import Path
import pypandoc
```

## Исходные данные

Мне доступны исходники [продуктовой документации](https://doc.ru.universe-data.ru/6.13.0-EE/index.html) продукта MDM 6.13 (Master Data Management) компании [Universe Data](https://universe-data.ru/). Документация построена с помощью Sphinx из файлов в разметке RST (ReStructuredText). 

Документация помимо текста содержит скрины из интерфейса MDM, исходники этих изображений так же доступны.

```python
DOCS_DIR = Path('../data/sphinx-service-docs-6.13.0-RU/docs/content')
START_GUIDE = DOCS_DIR / 'getting_started'
GUIDE = DOCS_DIR / 'guides'
IMAGES = DOCS_DIR / 'img'
```

```python
start_guide_pages = list(START_GUIDE.rglob('*.rst'))
guide_pages = list(GUIDE.rglob('*.rst'))
print(f'Всего страниц документации: {len(start_guide_pages) + len(guide_pages)}')
```

```python
images_paths = list(IMAGES.rglob('*.png'))
print(f'Всего изображений: {len(images_paths)}')
```

Разметка RST довольно избыточна, плохо понимается LLM и c ней неудобно работать. Для простоты конвертируем документы в GitHub Flavored Markdown.

```python
from tqdm import tqdm


converted_docs = Path('../data/markdown_docs')

for page_path in tqdm(start_guide_pages + guide_pages, desc='Converting pages', unit='page'):
    converted_page_path = (converted_docs / page_path.relative_to(DOCS_DIR)).with_suffix('.md')
    converted_page_path.parent.mkdir(parents=True, exist_ok=True)
    # print(converted_page_path)
    pypandoc.convert_file(page_path, format='rst', to="gfm", outputfile=converted_page_path)
    # break
```
